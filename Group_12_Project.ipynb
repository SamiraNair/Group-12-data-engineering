###**Import Libraries**

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###**Fetch Data from IMDb**

```python
url = "https://www.imdb.com/chart/top/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
}

response = requests.get(url, headers=headers)

if response.status_code == 200:

    soup = BeautifulSoup(response.content, 'html.parser')
    print("HTML content successfully parsed.")

else:
    print('Request failed with status code:', response.status_code)
```

###**EXTRACT**

**1. MOVIE IDS**

```python
movie_ids = []

movies = soup.find_all('li', class_='ipc-metadata-list-summary-item')

for movie in movies:
    movie_id_element = movie.find('a', class_='ipc-title-link-wrapper', href=True)

    if movie_id_element:
        # Extract the 'href' attribute which contains the movie ID
        movie_id_text = movie_id_element['href']

        # Use regex to extract the movie ID from the 'href'
        movie_id_match = re.search(r'/title/([^/?]+)', movie_id_text)

        # Check if the regex match is found
        if movie_id_match:
            # Extract the movie ID from the regex match
            movie_id = movie_id_match.group(1)
            movie_ids.append(movie_id)
            print(movie_id)
        else:
            print("Movie ID: Not found")
    else:
        print("Movie ID: Not found")
```

**2. MOVIE TITLES**

```python
movie_titles = []

movies= soup.find_all('li', class_ = 'ipc-metadata-list-summary-item')

for movie in movies:
    title_element = movie.find('h3', class_='ipc-title__text')
    if title_element:
        movie_title = title_element.get_text().strip()
        movie_titles.append(movie_title)
    else:
        movie_titles.append("Not found")

for title in movie_titles:
    print(title)
```

**3. MOVIE RATING**

```python
movie_ratings = []

movies = soup.find_all('li', class_='ipc-metadata-list-summary-item')

for movie in movies:
    rating_element = movie.find('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating')
    if rating_element:
        rating_text = rating_element['aria-label']
        print(rating_text)
        movie_ratings.append(rating_text)
    else:
        print("Not found")

for rating in movie_ratings:
    print(rating)
```

**4. MOVIE DURATION**

```python
duration = []

time = soup.findAll('span', class_='sc-935ed930-8 iLiQCS cli-title-metadata-item')

for div in time:
    text = div.text.strip()

    # Use a regular expression to check if the text matches the optional "Xh," "Xm," or "Xh Xm" pattern
    duration_match = re.match(r'(?:(\d+)h\s*)?(?:(\d+)m)?', text)

    # Extract hours and minutes if available
    hours = duration_match.group(1) if duration_match.group(1) else '0'
    minutes = duration_match.group(2) if duration_match.group(2) else '0'

    # Check if both hours and minutes are not zero
    if hours != '0' or minutes != '0':

        duration.append(f"{hours}h {minutes}m")
duration
```

**5. MOVIE VOTES**

```python
vote = []

votes_elements = soup.findAll('span', class_='ipc-rating-star--voteCount')

for element in votes_elements:
    vote_text = element.text
    numerical_value = vote_text.split('(')[1].split(')')[0]
    vote_count = f"{numerical_value}"
    vote.append(vote_count)

    if numerical_value == '0':
         print("Votes: Not found")
    else:
        print(vote_count)
```

**6. MOVIE YEAR**

```python
data_years = []

years = soup.findAll('span', class_='sc-935ed930-8 iLiQCS cli-title-metadata-item')

count = 1

for year in years:
    year_text = year.text
    extracted_years = re.findall(r'\b\d{4}\b', year_text)

    if extracted_years:
        # Concatenate the extracted years into a string
        years_string = ', '.join(extracted_years)
        data_years.append(years_string)
        print(f"{years_string}")
        count += 1
```

**7. MOVIE AGE RATING**

```python
data_age = []

age_rating = soup.find_all("span", {"class": "sc-935ed930-8 iLiQCS cli-title-metadata-item"})

count = 1

for age in age_rating:
    age_text = age.text
    # Remove text matching the duration pattern
    age_r = re.sub(r'(\d+h)?\s*(\d+m)?', '', age_text)
    age_r = re.sub(r'\b\d{4}\b', '', age_r)

    if age_r:
        data_age.append(age_r)
        print(f"{age_r}")
        count += 1
    elif count == 77:
        # 77th movie does not contain age rating, so print "NotRated"
        print(f"NotRated")
        count += 1
```

**8. MOVIE DIRECTORS**

```python
directors = []

for movie_id in movie_ids:
    url = f"https://www.imdb.com/title/{movie_id}/?ref_=chttp_t_1"

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        director_link = soup.find('a', class_='ipc-metadata-list-item__list-content-item')

        if director_link:
            director_name = director_link.text.strip()
            directors.append((movie_id, director_name))
        else:
            directors.append((movie_id, "Director name not found"))
    else:
        directors.append((movie_id, "Failed to fetch the webpage"))

for movie_id, director in directors:
    print(f"{director}")
```

**9. MOVIE REVIEWS**

```python
reviews = []

for movie_id in movie_ids:
    url = f"https://www.imdb.com/title/{movie_id}/reviews/?ref_=tt_ov_rt"

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")


        review_contents = soup.find_all('div', class_='content')

        if review_contents:

            first_review = review_contents[0].text.strip()
            reviews.append(first_review)

for review in reviews:
    print(review)
```

**Saved into a extract_movieData.csv as raw data**

```python
# Extract director names from the 'directors' list
director_names = [director[1] for director in directors]

# Create a DataFrame from the extracted data
data = {
    'Movie ID': pd.Series(movie_ids),
    'Movie Title': pd.Series(movie_titles),
    'IMDb Rating': pd.Series(movie_ratings),
    'Duration': pd.Series(duration),
    'Votes': pd.Series(vote),
    'Year': pd.Series(data_years),
    'Age Rating': pd.Series(data_age),
    'Director': pd.Series(director_names),
    'Review': pd.Series(reviews)
}

# Create DataFrame with NaN-filled columns
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('extract_movieData.csv', index=False)

# Display a message indicating that the CSV file has been created
print("Data has been successfully written to extract_movieData.csv")

df.head(10)
```

###**TRANSFORM**

**Step 1: DATA PREPROCESSING**

**1.Movie Title (Remove numbering, extra whitespace &convert to title case)**

```python
cleaned_movie_titles = []

for movie_title in movie_titles:
    # Remove numbering and extra whitespace, and convert to title case
    cleaned_movie_title = re.sub(r'^\d+\.\s*', '', movie_title).title()

    cleaned_movie_titles.append(cleaned_movie_title)

for title in cleaned_movie_titles:
    print(title)
```

**2.IMDb Rating (Extract only numerical value and format it to /10 )**

```python
cleaned_movie_ratings = []

for rating in movie_ratings:
    numerical_rating = re.search(r'(\d+\.\d+)', rating)
    if numerical_rating:
        formatted_rating = f"{numerical_rating.group(0)}/10"
        cleaned_movie_ratings.append(formatted_rating)

for rating in cleaned_movie_ratings:
    print(rating)
```

**3. Duration (Transform duration into minutes)**

```python
transformed_durations = []

for duration in duration:
    # Use a regular expression to extract hours and minutes
    duration_match = re.match(r'(\d+)h (\d+)m', duration)

    # Check if the match is successful
    if duration_match:
        # Extract hours and minutes if available
        hours = int(duration_match.group(1)) if duration_match.group(1) else 0
        minutes = int(duration_match.group(2)) if duration_match.group(2) else 0

        total_minutes = hours * 60 + minutes
        transformed_durations.append(total_minutes)
    else:
        # Handle the case where the pattern doesn't match
        transformed_durations.append(None)

# Print out the transformed durations
for duration in transformed_durations:
    print(duration)
```

**4. Votes (Transform the votes into numerical value Eg. 3k → 3000)**

```python
def transform_votes(vote_text):
    if vote_text.isdigit():
        return int(vote_text)
    elif 'K' in vote_text:
        return int(float(vote_text.replace('K', '')) * 1000)
    elif 'M' in vote_text:
        return int(float(vote_text.replace('M', '')) * 1000000)
    else:
        return None

transformed_votes = []

for vote_text in vote:
    numerical_vote = transform_votes(vote_text)
    transformed_votes.append(numerical_vote)

# Print out the transformed votes
for vote_value in transformed_votes:
    if vote_value is not None:
        print(f"{vote_value}")
    else:
        print("Votes: Not found")
```

**5. Age Rating (Handle missing age ratings by replacing them. As of the 249th movie, age ratings are available. For the 250th movie the age rating is not available (N/A).)To avoid this row being empty, we will input 'NotRated' into it.**

```python
while len(data_age) < 250:
    data_age.append('NotRated')

# Set the 250th entry to 'NotRated'
data_age[249] = 'NotRated'

print(len(data_age))
```

**6. Review (A couple of preprocessing step have taken for the review)**

*   Remove duplicate reviews.
*   Eliminate irrelevant words such as "Permalink" and the phrase "Was this review helpful. Sign in to vote." & something like "169 out of 177 found this helpful"
*   Remove special characters and whitespace.
*   Convert the text to lowercase.





```python
import re

cleaned_reviews = []

for review in reviews:
    # Remove duplicate reviews
    if review not in cleaned_reviews:
        # Remove irrelevant words and phrases
        cleaned_review = re.sub(r'Permalink', '', review)

        # Remove the specific phrase with count and "found this helpful"
        cleaned_review = re.sub(r'\b\d+\s*out\s*of\s*\d+\s*found\s*this\s*helpful\b', '', cleaned_review)


        # Eliminate the specific phrase with count and "Was this review helpful?"
        cleaned_review = re.sub(r'\s*Was\s*this\s*review\s*helpful\?[\s\S]*?vote\.\s*', '', cleaned_review)

        # Remove special characters and whitespace
        cleaned_review = re.sub(r'[^a-zA-Z0-9\s]', '', cleaned_review)

        # Convert the text to lowercase
        cleaned_review = cleaned_review.lower()

        cleaned_reviews.append(cleaned_review)

# Print out the cleaned reviews
for cleaned_review in cleaned_reviews:
    if cleaned_review.strip():
      # Remove the specific phrase with count and "found this helpful"
        cleaned_review = re.sub(r'\b\d+\s*out\s*of\s*\d+\s*found\s*this\s*helpful\b', '', cleaned_review)
        print(cleaned_review)
```

For your information, the columns 'Movie ID' and 'Year' underwent preprocessing during the extraction process. As for the 'Director' column, no preprocessing step was necessary, so there have been no changes to it.
